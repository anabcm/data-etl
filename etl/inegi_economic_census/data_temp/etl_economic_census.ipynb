```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://storage.googleapis.com/datamexico-data/inegi_economic_census/data_economic_census.csv", 
                 encoding="latin-1", 
                 low_memory=False)
```

```python
df1 = df
```

```python
df1.shape
```

```python
df1.head()
```

```python
df1 = df1.drop_duplicates(subset=["Año Censal", "Entidad", "Municipio", "Actividad Económica", "UE Unidades económicas"])
```

```python
df1["Municipio"]
```

```python
df1['Actividad Económica'].value_counts()
```

```python
df1.drop(['Unnamed: 118'], axis=1, inplace = True )
```

```python
temp = list(df1.columns[:4])
for i in range(4, len(df1.columns)):
    temp.append(df1.columns[i].split(' ')[0])
```

```python
df1.columns = temp
```

```python
df1.head()
```

```python
df1.drop( df1[ df1["Actividad Económica"] == "Total nacional" ].index , inplace=True)
df1.drop( df1[ df1["Actividad Económica"] == "Total estatal" ].index , inplace=True)
df1.drop( df1[ df1["Actividad Económica"] == "Total municipal" ].index , inplace=True)
#df1.drop( df1[ df1["Entidad"] == "00 Total Nacional" ].index , inplace=True)
```

```python
df1.head()
```

```python
df2 = df1
```

```python
df2['Municipio'].isnull().sum()
```

```python
df2 = df2[pd.notnull(df2["Municipio"])]
```

```python
df2['Municipio'].isnull().sum()
```

```python
df2 = df2.reset_index(drop=True)
```

```python
len(list(df2['Municipio'].unique()))
```

```python
list(df2['Entidad'].unique())
```

```python
piv_entidad = df2["Entidad"].str.split(" ", n=1, expand=True)
piv_municipio = df2["Municipio"].str.split(" ", n=1, expand=True)
df2["Entidad"] = piv_entidad[0] + piv_municipio[0]
```

```python
df2.rename(index=str, columns={"Año Censal":"year",
                              "Entidad":"mun_id",
                              "Actividad Económica":"actividad_economica"}, inplace=True)
```

```python
piv_actividad = df2["actividad_economica"].str.split(" ", n=1, expand=True)
df2["actividad_economica"] = piv_actividad[0]
```
