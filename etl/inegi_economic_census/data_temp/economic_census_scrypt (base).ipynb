```python
df = pd.read_csv("SAIC201906071109546.csv", encoding="latin-1", header=4)
```

```python
df.drop(['Unnamed: 127'], axis=1, inplace = True )
```

```python
type(list(df.columns))
```

```python
temp = list(df.columns[:4])
for i in range(4, len(df.columns)):
    temp.append(df.columns[i].split(' ')[0])
```

```python
len(temp)
```

```python
df.columns = temp
```

```python
df
```

```python
df.drop( df[ df['Actividad Económica'] == "Total nacional" ].index , inplace=True)
df.drop( df[ df['Actividad Económica'] == "Total estatal" ].index , inplace=True)
df.drop( df[ df['Actividad Económica'] == "Total municipal" ].index , inplace=True)
```

```python
df
```

```python
df.drop( df[ df['Entidad'] == "00 Total Nacional" ].index , inplace=True)
```

```python
df.drop(df.tail(3).index,inplace=True)
```

```python
df['Municipio'].isnull().sum()
```

```python
df = df[pd.notnull(df["Municipio"])]
```

```python
df
```

```python
piv_entidad = df["Entidad"].str.split(" ", n=1, expand=True)
piv_municipio = df["Municipio"].str.split(" ", n=1, expand=True)

df["Entidad"] = piv_entidad[0] + piv_municipio[0]
```

```python
df
```

```python
df.rename(index=str, columns={"Año Censal":"year",
                              "Entidad":"mun_id",
                              "Actividad Económica":"actividad_economica"}, inplace=True)
```

```python
df
```

```python
piv_actividad = df["actividad_economica"].str.split(" ", n=1, expand=True)
df["actividad_economica"] = piv_actividad[0]
```

```python
df.drop(['Municipio'], axis=1, inplace = True )
```
