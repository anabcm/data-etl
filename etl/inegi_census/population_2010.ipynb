```python
import pandas as pd
from simpledbf import Dbf5
```

```python
pd.options.display.max_columns = 96
```

```python
# Loading new columns labels (2010) to the file
path_labels = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQrgV5L1KtlO35d4-j2uUEHifL_S2NNmqM9dU6SuMCbHfS_tYQxaTwJJmOaFz-k9kemwIlfN9i381lt/pub?output=xlsx"
df_labels = pd.ExcelFile(path_labels)
```

```python
dbf = Dbf5("Personas_01.dbf", codec="latin-1")
df = dbf.to_dataframe()
```

```python
df.columns = df.columns.str.lower()
```

```python
# Adding IDs columns and renaming factor as population
df["loc_id"] = df["ent"] + df["mun"] + df["loc50k"]
df["mun_id_trab"] = df["ltrabpai_c"] + df["ltrabmun_c"]
df["year"] = 2010
df["edad"] = df["edad"].astype(int)
```

```python
df.loc[df["mun_id_trab"] > 33000, "mun_id_trab"] = 0
df["mun_id_trab"].replace(pd.np.nan , 0, inplace=True)
```

```python
params_nan = ["laboral_condition", "time_to_work", "transport_mean_work", "time_to_ed_facilities", "transport_mean_ed_facilities"]
```

```python
# Transforming certains str columns into int values
df["loc_id"] = df["loc_id"].astype(int)
df["factor"] = df["factor"].astype(int)
```

```python
df.head()
```

```python
params = ["sexo", "parent", "sersalud", "dhsersal1", "nivacad"]
params_nan = ["laboral_condition", "time_to_work", "transport_mean_work", "time_to_ed_facilities", "transport_mean_ed_facilities"]
params_int = ["loc_id", "year", "mun_id_trab"]
```

```python
# For cycle in order to change the content of a column from previous id, into the new ones (working for translate too)
for sheet in params:
    df_l = pd.read_excel(df_labels, sheet)
    df[sheet] = df[sheet].astype(int)
    df[sheet] = df[sheet].replace(dict(zip(df_l.prev_id, df_l.id)))
```

```python
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQrgV5L1KtlO35d4-j2uUEHifL_S2NNmqM9dU6SuMCbHfS_tYQxaTwJJmOaFz-k9kemwIlfN9i381lt/pub?output=xlsx"
pivote = pd.read_excel(url, sheet_name="tramo_edad", encoding="latin-1", dtype={"interval_upper": "int64", "interval_lower": "int64"})
for años in df.edad.unique():
    for level in range(pivote.shape[0]):
        if (años >= pivote.interval_lower[level]) & (años < pivote.interval_upper[level]):
            df.edad.replace(años, str(pivote.id[level]), inplace=True)
            break

df.edad = df.edad.astype("int")
```

```python
df = df.groupby(params + params_nan + params_int + ["edad"]).sum().reset_index(col_fill="ffill")
```

```python
# Renaming of certains columns
df.rename(index=str, columns={"factor": "population",
                              "edad": "age",
                              "sexo": "sex",
                              "nivacad": "academic_degree"}, inplace=True)
```
